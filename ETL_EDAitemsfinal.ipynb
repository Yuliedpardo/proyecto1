```python
import pandas as pd
import ast
import numpy as np
from ast import literal_eval
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob
import re
from datetime import datetime
from dateutil import parser
```

leemos el archivo json

```python
input_file = 'australian_users_items.json'

# Leer el contenido del archivo JSON
with open(input_file, 'r', encoding='utf-8') as f:
    data = f.readlines()

# Convertir las líneas a registros JSON
records = [eval(line.strip()) for line in data]

# Crear el DataFrame a partir de los registros
df_items= pd.DataFrame(records)
```

```python
df_items
```

desanidamos la columna items

```python
unpacked_data = []
for index, row in df_items.iterrows():
    user_id = row['user_id']
    user_url = row['user_url']
    user_items_count = row['items_count']
    for df_items_dict in row['items']:
        df_items_dict['user_id'] = user_id
        df_items_dict['user_url'] = user_url
        df_items_dict['items_count'] = user_items_count
        unpacked_data.append(df_items_dict)

df_items = pd.DataFrame(unpacked_data)
```

```python
df_items
```

guardamos el archivo a csv

```python
df_items.to_csv('df_itemsdesanidado2.csv', index=False)
```
